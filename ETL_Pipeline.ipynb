```python
# Step 6: Fully Automated ETL Pipeline with Summary & Timestamp
from pipeline import extract_data, transform_data
from logger import setup_logger
import os
from datetime import datetime
import matplotlib.pyplot as plt

# Initialize logger
logger = setup_logger()

# 1️⃣ Define file paths
raw_path = "data/raw/raw_data.csv"
processed_dir = "data/processed"

# Create processed folder if it doesn't exist
os.makedirs(processed_dir, exist_ok=True)

# Timestamped processed file
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
processed_file = os.path.join(processed_dir, f"processed_data_{timestamp}.csv")

logger.info("===== ETL Pipeline Started =====")

# 2️⃣ Extract data
df = extract_data(raw_path, logger)

# 3️⃣ Transform data
processed_df = transform_data(df, logger)

# 4️⃣ Save transformed data
processed_df.to_csv(processed_file, index=False)
logger.info(f"Processed data saved to {processed_file}")

# 5️⃣ Automatic summary report
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
summary = {
    "num_rows": processed_df.shape[0],
    "num_columns": processed_df.shape[1],
    "numeric_columns": numeric_cols,
    "categorical_columns": categorical_cols
}
logger.info(f"Dataset Summary: {summary}")

# 6️⃣ Optional: Visualizations
# Numeric columns histograms
for col in numeric_cols:
    plt.figure()
    processed_df[col].hist(bins=10)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Categorical columns bar plots (one-hot encoded)
for col in processed_df.columns:
    if col not in numeric_cols:
        plt.figure()
        processed_df[col].value_counts().plot(kind='bar')
        plt.title(f"Counts of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()

logger.info("===== ETL Pipeline Completed Successfully =====")

# Display first 5 rows
processed_df.head()
```
